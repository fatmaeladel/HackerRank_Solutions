```python
#!/bin/python3

import math
import os
import random
import re
import sys
def solve(s):
    return re.sub(r'\b[a-zA-Z]', lambda x: x.group().upper(), s)

if __name__ == '__main__':
    # fptr = open(os.environ['OUTPUT_PATH'], 'w') # This line caused the error

    s = input()

    result = solve(s)

    # fptr.write(result + '\n') # This line is also not needed for console output
    # fptr.close() # This line is also not needed for console output
    print(result)
```
