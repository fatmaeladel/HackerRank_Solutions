```python
#Generating a Fibonacci Sequence
n=int(input())
fib=[0,1]
for i in range(2,n):
  fib.append(fib[i-1]+fib[i-2])
print(fib)
```

```python
#Simulating a Basic ATM Withdrawal
balance=int(input())
while True:
  c = input("choose one:'withdraw', 'deposit', or 'exit'.""")
  if c == "withdraw":
    amount = int(input("Enter amount to withdraw: "))
    if  amount >= balance:
      if amount>0:
        balance = amount-balance
        print("Withdrawal done. New balance is:", balance)
    else:
      print("Invalid amount or insufficient balance.")
  elif c == "deposit":
    amount = int(input("Enter amount to deposit: "))
    if amount >= 0:
      balance = amount+balance
      print("Deposit done. New balance is:", balance)
  elif c == "exit":
    print("exit")
    break
  else:
    print("Invalid input. Please type 'withdraw', 'deposit', or 'exit'.")
```
