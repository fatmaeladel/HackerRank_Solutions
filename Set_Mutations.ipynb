```python
a=int(input())
A=set(map( int ,input().split()))
n=int(input())
for _ in range (n):
    e , _=input().split()
    E=set(map(int ,input().split()))
    if e =="intersection_update":
        A.intersection_update(E)
    elif e=="difference_update":
        A.difference_update(E)
    elif e=="symmetric_difference_update":
        A.symmetric_difference_update(E)
    elif e=="update":
        A.update(E)
print(sum(A))
```
