```python
N=int(input())
Stu=[]
for _ in range (N):
    name=input()
    grade=float(input())
    Stu.append([name , grade])
grades=[grade for name , grade in Stu]
grades_sorted=sorted(set(grades))
secl=grades_sorted[1]
secl_name=[name for name , grade in Stu if grade == secl]
secl_name.sort()
for name in secl_name:
    print(name)

```
