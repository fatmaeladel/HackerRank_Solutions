```python
def average(array):
    arr1=set(arr)
    avg= sum(set(arr1))/len(arr1)
    return float(avg)

if __name__ == '__main__':
    n = int(input())
    arr = list(map(int, input().split()))
    result = average(arr)
    print(result)
```
